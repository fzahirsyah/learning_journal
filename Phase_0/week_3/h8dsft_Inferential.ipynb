## Hypotesis Testing

H0 = Provinsi Jawa Barat, Jawa Timur, dan Jawa Tengah memiliki rata-rata kesembuhan yang signifikan berbeda

H1 = Provinsi Jawa Barat, Jawa Timur, dan Jawa Tengah memiliki rata-rata kesembuhan yang Tidak signifikan berbeda

```python
jawa_df = covid_df.query('Location == "Jawa Barat" or Location == "Jawa Timur" or Location == "Jawa Tengah"')
jawa_df.head()
```

```python
jawa_barat = jawa_df[jawa_df['Location'] == 'Jawa Barat'].groupby(jawa_df['Date']).sum()['Total Recovered']
jawa_tengah = jawa_df[jawa_df['Location'] == 'Jawa Tengah'].groupby(jawa_df['Date']).sum()['Total Recovered']
jawa_timur = jawa_df[jawa_df['Location'] == 'Jawa Timur'].groupby(jawa_df['Date']).sum()['Total Recovered']

print('Rata-rata kesembuhan Provinsi Jawa Barat  : ',jawa_barat.mean())
print('Rata-rata kesembuhan Provinsi Jawa Tengah : ',jawa_tengah.mean())
print('Rata-rata kesembuhan Provinsi Jawa Timur  : ',jawa_timur.mean())
```

```python
f_stat,p_value = stats.f_oneway(jawa_barat,jawa_tengah,jawa_timur)
print('P-value:',p_value)
```

Karena P-value kurang dari 0.05 maka H0 gagal ditolak sehingga Provinsi Jawa Barat, Jawa Timur, dan Jawa Tengah memiliki rata-rata kesembuhan yang signifikan berbeda
