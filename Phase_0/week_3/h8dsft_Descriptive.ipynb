### 1. Finding measure of tendency from the dataset

```python
mean = round(covid_df[num_columns].mean().rename('Mean'),2)
median = round(covid_df[num_columns].median().rename('Median'),2)
mode = covid_df[num_columns + obj_columns].mode(dropna=True).iloc[0].rename('Mode')


central_tendency = pd.concat([mean, median,mode], axis=1)
central_tendency
```

Table above giving information about central tendency for every columns inside the dataset. Theres value that giving information about numerical data for example New Cases, New Deaths, New Recovered etc. These column giving us a valid central tendency. Different cases with Longitude and Latitude. These two columns are categorical which means that mean, median, and mode is just giving false information. Because Latitute or Longitude doesn't have any central tendency.

### 2. Find Measure of Variance from the dataset

```python
variance = covid_df[num_columns].var().rename('Variance')
stdev = covid_df[num_columns].std().rename('Standard Deviation')


Measure_of_variance = pd.concat([variance, stdev], axis=1)
Measure_of_variance
```

```python
np.argmax(Measure_of_variance['Standard Deviation'])
```

Table above summarize about variance and standard deviation for each numbers columns. most of them are within the range of +3 which is good because the data more likely to become normal distribution. The most variance in standard deviation is New Deaths and Total Regencies, both columns are having more spread data because their standard deviation is 0.764 and 0.687 respectively.
