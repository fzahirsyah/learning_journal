## Import Data

```python
import pandas as pd
import numpy as np
import requests
from bs4 import BeautifulSoup 
from selenium import webdriver


%matplotlib inline
```

```python
driver = webdriver.Chrome('chromedriver.exe')

url="https://www.imdb.com/search/title/?genres=animation"
driver.get(url)
html = driver.page_source

soup = BeautifulSoup(html, "html.parser")
print(soup.prettify()[:700])
```

```python
soup.find_all( 'div', {"class":"inline-block ratings-imdb-rating"}, "strong" )
```

```python
data = pd.DataFrame()

data['Title'] = [ title.get_text() for title in soup.find_all("a",{'href':"/title/tt2560140/?ref_=adv_li_tt"}) ]
data['Rating'] = [ rating.get_text() for rating in soup.find_all( 'div', {"class":"inline-block ratings-imdb-rating"} ) ]
data['Description'] = [ description.get_text() for description in soup.find_all( 'p', {"class":"formats-price"} ) ]
data['Votes'] = [ votes.get_text() for votes in soup.find_all( 'p', {"class":"formats-price"} ) ]
data['Poster'] = [ img['src'] for img in soup.find_all( 'img',{"class":"product-list-img"} ) ]
```
