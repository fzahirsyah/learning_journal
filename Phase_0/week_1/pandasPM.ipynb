# Query

```python
df.head()
```

```python
df[df['year_id']>2010]['pts']
```

```python
# year id >2010
# fran_d Warriors
# pts diatas 100
# game result w


df[
    (df['year_id']>2010) &
    (df['year_id']<2013) &
    (df['fran_id']=='Warriors') &
    (df['pts']>=100) &
    (df['game_result']=='W')
][['year_id','fran_id','pts','opp_pts','opp_fran','game_result']]
```

```python
# year id >2010
# fran_d Warriors
# pts diatas 100
# game result w

df.query("year_id > 2013 & fran_id == 'Warriors' & pts >= 100 & game_result == 'W'")[['year_id','fran_id','pts','opp_pts','opp_fran','game_result']]
```

```python
# create new columns
df_new = df.copy()
df_new.head()
```

```python
df_new['selisih_point'] = abs(df_new['pts'] - df_new['opp_pts'])
df_new.head()
```

```python
df_new.rename(columns={'game_location':'game_loc',
                        'game_result':'result'},inplace=True)
# inplace digunakan untuk mereplace dataframe
df_new
```

```python
# drop Columns
df_new.drop(['elo_i','elo_n'],axis=1,inplace=True)
```
